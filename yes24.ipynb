```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html = urlopen("http://www.yes24.co.kr/24/category/bestseller")
bsObject = BeautifulSoup(html, "html.parser")
bsObject
```

```python
for n in range(0,40):
    p = bsObject.find("p", {"id":"location_"+str(n)})
    title = p.select('img')[0].get('alt')
    author = p.next_sibling.next_sibling.next_sibling.next_sibling.select('a')[0].get_text()
    publisher = p.next_sibling.next_sibling.next_sibling.next_sibling.get_text().split()[-1]
    print(n+1, title, author, publisher)
```

```python
import csv
with open('예스24도서 베스트셀러 순위.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['순위', '제목', '저자', '출판사'])
    
    for n in range(0,40):
     p = bsObject.find("p", {"id":"location_"+str(n)})
     title = p.select('img')[0].get('alt')
     author = p.next_sibling.next_sibling.next_sibling.next_sibling.select('a')[0].get_text()
     publisher = p.next_sibling.next_sibling.next_sibling.next_sibling.get_text().split()[-1]
     writer.writerow([n+1, title, author, publisher])
```
